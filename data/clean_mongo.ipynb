```python
import pandas as pd
```

```python
df_raw = pd.read_csv('beer_reviews.csv')
df_raw.columns
```

```python
df = df_raw[[ 'review_overall',
       'review_aroma', 'review_appearance', 'review_profilename', 'beer_style',
       'review_palate', 'review_taste', 'beer_name',
       'beer_beerid']]
```

```python
df_na = df
beer_count = df_na.groupby(['beer_name', 'beer_beerid']).count()['review_overall']
beer_avg = df_na.groupby(['beer_name', 'beer_beerid']).mean()
beer_avg['review_count'] = beer_count
```

```python
beer_df = beer_avg.reset_index()
beer_clean = beer_df.set_index('beer_name')
```

```python
sorted_df = beer_clean.sort_values(by = 'review_count', ascending= False)
beer_feature_count = sorted_df.groupby(['beer_name', 'beer_beerid']).count()['review_aroma']
```

```python
df_na.groupby(['beer_name', 'beer_beerid']).count()['review_aroma'].sort_values(ascending = False)
```

```python
beer_table = df_raw.groupby(['beer_name', 'beer_beerid']).mean()[['brewery_id','review_overall',	'review_aroma',	'review_appearance',	'review_palate',	'review_taste']]
beer_count = df_raw.groupby(['beer_name', 'beer_beerid']).count()['review_overall']
beer_table['review_count'] = beer_count
```

```python
beer_1000 = beer_table.sort_values(by = 'review_count', ascending = False).head(10000)
beer_reset = beer_1000.reset_index()
beer_df = beer_reset
beer_df['Beer_name'] = beer_reset['beer_name']
beer_table = beer_df.set_index('Beer_name')
```

```python
brewery_table = df_raw.groupby(['brewery_id']).mean()[['review_overall',	'review_aroma',	'review_appearance',	'review_palate',	'review_taste']]
brewery_count = df_raw.groupby(['brewery_id']).count()['review_overall']
brewery_table['review_count'] = brewery_count
brewery_df = brewery_table.sort_values(by = 'review_count', ascending = False)
brewery_df
```

```python
aroma_table = beer_df.sort_values(by=['review_aroma', 'review_count'], ascending = [False,False]).head(10000).set_index('beer_name')
appearance_table = beer_df.sort_values(by=['review_appearance', 'review_count'], ascending = [False,False]).head(10000).set_index('beer_name')
palate_table = beer_df.sort_values(by=['review_palate', 'review_count'], ascending = [False,False]).head(10000).set_index('beer_name')
taste_table = beer_df.sort_values(by=['review_taste', 'review_count'], ascending = [False,False]).head(10000).set_index('beer_name')
taste_table
```

```python
from pymongo import MongoClient
from config import mongo_uri

collection = MongoClient(mongo_uri).beer_db.beer_table
collection.insert_one(beer_table.to_dict())
```

```python
collection = MongoClient(mongo_uri).beer_db.feature_tables
collection.insert_many([aroma_table.to_dict(), appearance_table.to_dict(), palate_table.to_dict(), taste_table.to_dict()])
```
